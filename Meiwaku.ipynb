```python
import json
import os
from collections import Counter
```

```python
all_chars = []
for file in os.listdir("crawled"):
    filepath = os.path.join("crawled", file)
    chars = json.load(open(filepath))
    all_chars.extend(chars['data'])
```

```python
all_tropes = {}
for i,c in enumerate(all_chars):
    for t in c['tropes']:
        if t['trope_type'] in all_tropes:
            all_tropes[t['trope_type']].append((i,c['name']))
        else:
            all_tropes[t['trope_type']] = [(i,c['name'])]
print('total_tropes len',len(all_tropes))
```

```python
# for t in all_tropes:
#     print (t, len(all_tropes[t]))
selected_tropes = sorted([(t,len(all_tropes[t])) for t in all_tropes],key= lambda x:-x[1])[:220]
with open('selected_tropes.txt', 'w+') as f:
    for name,time in selected_tropes:
        f.write(f"{name}\n")
```

```python
def find_chars(query):
    res = []
    for i,c in enumerate(all_chars):
        if query.lower() in c['name'].lower():
            res.append((c['name'],i))
    return res
# find_chars("yukino")
```

```python
def similar_chars(char_idx_to_find, up_lim=10):
    similar_char = {}
    trope_count = len(all_chars[char_idx_to_find]["tropes"])
    print(f'finding {all_chars[char_idx_to_find]["name"]} with total of {trope_count} trope')
    for item in all_chars[char_idx_to_find]['tropes']:
        trope_type = item['trope_type']
        for char_idx,name in all_tropes[trope_type]:
            if char_idx in similar_char:
                similar_char[char_idx].append(trope_type)
            else:
                similar_char[char_idx] = [trope_type]
    return sorted([(t, len(set(similar_char[t])) , len(set(similar_char[t])) / trope_count,set(similar_char[t]))   for t in similar_char], key=lambda x:-x[1])[1:up_lim]
```

```python
all_tropes_desc = {}
for c in all_chars:
    for t in c['tropes']:
        if t['trope_type'] in all_tropes_desc:
            all_tropes_desc[t['trope_type']].append(t['trope_description'])
        else:
            all_tropes_desc[t['trope_type']] = [t['trope_description']]
print('total_tropes len',len(all_tropes_desc))
```

```python
with open('selected_tropes.txt','r') as f:
    selected_tropes = [line[:-1] for line in f.readlines()]

for trope in selected_tropes:
    with open(f'wac/{trope}.txt','w+') as f:
        for sent in all_tropes_desc[trope]:
            f.write(sent+'\n')

with open(f'wac/{trope}.txt','w+') as f:
    for trope in selected_tropes:
        for sent in all_tropes_desc[trope]:
            f.write(sent+'\n')
```

```python
import spacy

nlp = spacy.load("en_core_web_md")
```

```python
from spacy.tokenizer import Tokenizer
tokenizer = Tokenizer(nlp.vocab)
import re
```

```python
def preprocess_sent(sent):
    sent = re.sub("[^A-z0-9 ']", "",sent)
    sent = sent.lower()
    sent = re.sub('\d+',"<NUM>", sent)
    sent = tokenizer(sent)
    return sent
preprocess_sent("She frequently derides 69 Sena's 'cow udders', which are the source of her nickname 'Meat'. One of their early arguments had")
```

```python

for i, trope in enumerate(selected_tropes):
    with open(f'data.txt','a+') as f:
        for sent in all_tropes_desc[trope]:
            sent = preprocess_sent(sent)
            if sent:
                f.write(f"{i},{sent}\n")
```
